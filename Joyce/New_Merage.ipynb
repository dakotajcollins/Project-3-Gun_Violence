```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
Gun_MS_2019 = "Resource/MASS_SHOOTINGS_IN_2019.csv"
Gun_MS_2020 = "Resource/MASS_SHOOTINGS_IN_2020.csv"
Gun_MS_2021 = "Resource/MASS_SHOOTINGS_IN_2021.csv"
Statelatlon="Resource/US_City_latLon.csv"
```

```python
file_one_df = pd.read_csv(Gun_MS_2019)
file_two_df = pd.read_csv(Gun_MS_2020)
file_three_df = pd.read_csv(Gun_MS_2021)
LatLon_df=pd.read_csv(Statelatlon)
```

```python
columnsplit=file_one_df['Incident Date'].str.split(",", expand=True)
file_one_df=file_one_df.assign(Date=columnsplit[0],Year=columnsplit[1])
file_one_df_1=file_one_df.drop(columns=["Incident Date","Operations"])
file_one_df_1.head()
```

```python
columnsplit_2=file_two_df['Incident Date'].str.split(",", expand=True)
file_two_df=file_two_df.assign(Date=columnsplit_2[0],Year=columnsplit_2[1])
file_two_df_2=file_two_df.drop(columns=["Incident Date", "Operations"])
file_two_df_2.head()
```

```python
columnsplit_3=file_three_df['Incident Date'].str.split(",", expand=True)
file_three_df=file_three_df.assign(Date=columnsplit_3[0],Year=columnsplit_3[1])
file_three_df_3=file_three_df.drop(columns=["Incident Date", "Operations"])
file_three_df_3.head()
```

```python
# LatLon
LatLon_df.head()
```

```python
columnsplit_4=LatLon_df['Place Name'].str.split(",", expand=True)
LatLon_df=LatLon_df.assign(City=columnsplit_4[0],County=columnsplit_4[1])
LatLon_df=LatLon_df.drop(columns=["County", "Place Name"])
LatLon_df.head()
```

```python
Gun_map_2019=file_one_df_1.merge(LatLon_df, left_on='City Or County', right_on='City')
Gun_map_2020=file_two_df_2.merge(LatLon_df, left_on='City Or County', right_on='City')
Gun_map_2021=file_three_df_3.merge(LatLon_df, left_on='City Or County', right_on='City')
Gun_map_2019.head(50)
```

```python
Gun_map_2019=Gun_map_2019.drop(columns=["City"])
Gun_map_2020=Gun_map_2020.drop(columns=["City"])
Gun_map_2021=Gun_map_2021.drop(columns=["City"])
Gun_map_2019
```

```python
Gun_map_2019.to_csv("Resource/Gun_violence_2019.csv",index=False)
Gun_map_2020.to_csv("Resource/Gun_violence_2020.csv",index=False)
Gun_map_2021.to_csv("Resource/Gun_violence_2021.csv",index=False)
```
