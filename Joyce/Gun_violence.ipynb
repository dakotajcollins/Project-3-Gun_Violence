```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
Gun_MS_2019 = "MASS_SHOOTINGS_IN_2019.csv"
Gun_MS_2020 = "MASS_SHOOTINGS_IN_2020.csv"
Gun_MS_2021 = "MASS_SHOOTINGS_IN_2021.csv"
Statelatlon="statelatlong.csv"
```

```python
file_one_df = pd.read_csv(Gun_MS_2019)
file_two_df = pd.read_csv(Gun_MS_2020)
file_three_df = pd.read_csv(Gun_MS_2021)
LatLon_df=pd.read_csv(Statelatlon)
```

```python
#2019
file_one_df.head(100)
```

```python
columnsplit=file_one_df['Incident Date'].str.split(",", expand=True)
file_one_df=file_one_df.assign(Date=columnsplit[0],Year=columnsplit[1])
file_one_df.head()
```

```python
file_one_df_1=file_one_df.drop(columns=["Incident Date","Operations"])
file_one_df_1
```

```python
# 2020
file_two_df.head(100)
```

```python
columnsplit_2=file_two_df['Incident Date'].str.split(",", expand=True)
file_two_df=file_two_df.assign(Date=columnsplit_2[0],Year=columnsplit_2[1])
file_two_df.head()
```

```python
file_two_df_2=file_two_df.drop(columns=["Incident Date", "Operations"])
file_two_df_2
```

```python
# 2021
file_three_df.head(100)
```

```python
columnsplit_3=file_three_df['Incident Date'].str.split(",", expand=True)
file_three_df=file_three_df.assign(Date=columnsplit_3[0],Year=columnsplit_3[1])
file_three_df.head()
```

```python
file_three_df_3=file_three_df.drop(columns=["Incident Date", "Operations"])
file_three_df_3
```

```python
# LatLon
LatLon_df.head()
```

```python
#Join lat & lon

Gun_map_2019=file_one_df_1.merge(LatLon_df, left_on='State', right_on='City')
Gun_map_2020=file_two_df_2.merge(LatLon_df, left_on='State', right_on='City')
Gun_map_2021=file_three_df_3.merge(LatLon_df, left_on='State', right_on='City')
Gun_map_2019.head(50)
```

```python
# Final DataFrame

Gun_violence_2019=Gun_map_2019[["Incident ID","State_x","City Or County","Address",
                               "# Killed", "# Injured","Date","Year", "State_y", "Latitude","Longitude"]]
Gun_violence_2019=Gun_violence_2019.rename(columns={"Incident ID" : "ID",
                                                    "State_x":"State",
                                                    "City Or County":"City",
                                                    "# Killed":"#_Killed",
                                                    "# Injured":"#_Injured",
                                                   "State_y": "USPS",
                                                   "Latitude":"Lat",
                                                   "Longitude":"Lon"})
Gun_violence_2019.head()
```

```python
Gun_violence_2019.to_csv("Gun_violence_2019.csv",index=False)
```

```python
Gun_violence_2020=Gun_map_2020[["Incident ID","State_x","City Or County","Address",
                               "# Killed", "# Injured","Date","Year", "State_y", "Latitude","Longitude"]]
Gun_violence_2020=Gun_violence_2020.rename(columns={"Incident ID" : "ID",
                                                    "State_x":"State",
                                                    "City Or County":"City",
                                                    "# Killed":"#_Killed",
                                                    "# Injured":"#_Injured",
                                                   "State_y": "USPS",
                                                   "Latitude":"Lat",
                                                   "Longitude":"Lon"})
Gun_violence_2020.head()
```

```python
Gun_violence_2020.to_csv("Gun_violence_2020.csv",index=False)
```

```python
Gun_violence_2021=Gun_map_2021[["Incident ID","State_x","City Or County","Address",
                               "# Killed", "# Injured","Date","Year", "State_y", "Latitude","Longitude"]]
Gun_violence_2021=Gun_violence_2021.rename(columns={"Incident ID" : "ID",
                                                    "State_x":"State",
                                                    "City Or County":"City",
                                                    "# Killed":"#_Killed",
                                                    "# Injured":"#_Injured",
                                                   "State_y": "USPS",
                                                   "Latitude":"Lat",
                                                   "Longitude":"Lon"})
Gun_violence_2021.head()
```

```python
Gun_violence_2021.to_csv("Gun_violence_2021.csv",index=False)
```
